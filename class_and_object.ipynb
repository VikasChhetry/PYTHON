```python
class Studant:
    school = "HCS"
    def __init__(self,name,age,classs,dob):
        self.name = name
        self.age = age
        self.classs = classs
        self.dob = dob
    @classmethod    #classmethod to change the school name
    def school_name(cls,name):
        cls.school= name

    def display(self):
        print ("School name is: {}\n Studant name is: {}\n Age is: {}\n Class name is: {}\n DOB is: {}\n".format(self.school,self.name,self.age,self.classs,self.dob))    

obj1=Studant("Vikas",20,11,"9/12/2003")
obj1.display()
Studant.school_name("RMP")  # Change the school_name using the class method
obj1.display()  #display again to see that
```
