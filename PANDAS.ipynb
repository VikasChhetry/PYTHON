```python
import tkinter as tk
from math import *

# Function to handle button click
def click(event):
    text = event.widget.cget("text")
    if text == "=":
        try:
            result = str(eval(display.get()))
            display.delete(0, tk.END)
            display.insert(tk.END, result)
        except Exception as e:
            display.delete(0, tk.END)
            display.insert(tk.END, "Error")
    elif text == "C":
        display.delete(0, tk.END)
    else:
        display.insert(tk.END, text)

# Create main window
win = tk.Tk()
win.title("Scientific Calculator")
win.geometry("400x600")

# Create display
display = tk.Entry(win, font=("Arial", 20), bd=10, insertwidth=2, width=14, borderwidth=4)
display.grid(row=0, column=0, columnspan=4)

# Define buttons
buttons = [
    "7", "8", "9", "/",
    "4", "5", "6", "*",
    "1", "2", "3", "-",
    ".", "0", "=", "+",
    "sin", "cos", "tan", "log",
    "sqrt", "(", ")", "C"
]

# Add buttons to the window
row = 1
col = 0
for button in buttons:
    b = tk.Button(win, text=button, padx=20, pady=20, font=("Arial", 14))
    b.grid(row=row, column=col)
    b.bind("<Button-1>", click)
    col += 1
    if col > 3:
        col = 0
        row += 1

# Run the main loop
win.mainloop()
```

```python
import matplotlib.pyplot as plt

# Define parameters
u = 5  # initial velocity in meters per second
a = 2  # acceleration in meters per second squared
time_steps = 100  # number of time steps
time_end = 10  # end time in seconds

# Create lists to store time and velocity values
t_values = []
v_values = []

# Calculate time step increment
delta_t = time_end / time_steps

# Compute values using a for loop
for step in range(time_steps + 1):
    t = step * delta_t
    v = u + a * t
    t_values.append(t)
    v_values.append(v)

# Create the plot
plt.plot(t_values, v_values, label='v = u + at')

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs. Time for v = u + at')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
